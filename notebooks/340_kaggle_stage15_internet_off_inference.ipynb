# Stage 15: Kaggle Internet-OFF inference

SettingsでInternetをOFF、AcceleratorをGPUにし、Stage 15 package DatasetをAdd Inputしてから全セル実行します。出力CSVは1個だけです。

```python
from pathlib import Path
import json, subprocess, sys
input_root = Path('/kaggle/input')
manifests = [p for p in input_root.rglob('manifest.json') if 'stage15' in p.read_text(errors='ignore')]
assert len(manifests) == 1, f'Expected one Stage 15 package manifest, found: {manifests}'
package_dir = manifests[0].parent
samples = list(input_root.rglob('sample_submission.csv'))
competition_samples = [p for p in samples if (p.parent / 'test').is_dir()]
assert len(competition_samples) == 1, f'Competition data not found uniquely: {competition_samples}'
data_dir = competition_samples[0].parent
sys.path.insert(0, str(package_dir / 'src'))
print('package:', package_dir)
print('competition data:', data_dir)
```

```python
from rogii.cli.stage15_infer import main
output = Path('/kaggle/working/submission.csv')
main(['--package-dir', str(package_dir), '--data-dir', str(data_dir), '--output', str(output), '--device', 'auto'])
```

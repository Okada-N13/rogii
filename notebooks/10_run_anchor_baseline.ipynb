# Last-known TVT anchor baseline

Run `00_colab_setup.ipynb` first in the same Colab runtime. Start with the 8-well smoke test, then run all wells with a new run ID.

```python
from pathlib import Path
import json
import os
import subprocess

repo_dir = Path('/content/ROGII')
assert repo_dir.is_dir(), 'Run 00_colab_setup.ipynb first'
os.chdir(repo_dir)
assert 'ROGII_DATA_DIR' in os.environ
assert 'ROGII_ARTIFACT_DIR' in os.environ
```

```python
RUN_ID = 'baseline_anchor_smoke_v001'
subprocess.run([
    'uv', 'run', 'rogii-train',
    '--config', 'configs/experiment/baseline_anchor.yaml',
    '--run-id', RUN_ID,
    '--data-dir', os.environ['ROGII_DATA_DIR'],
    '--artifact-dir', os.environ['ROGII_ARTIFACT_DIR'],
    '--limit-wells', '8',
], check=True)
```

```python
run_dir = Path(os.environ['ROGII_ARTIFACT_DIR']) / RUN_ID
metrics = json.loads((run_dir / 'metrics.json').read_text())
metrics
```

## Full run

After the smoke test succeeds, change `RUN_ID`, remove `--limit-wells 8`, and rerun the training cell. A run directory is never overwritten.
